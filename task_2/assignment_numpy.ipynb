* __Задача 1__: Подсчитать произведение ненулевых элементов на диагонали прямоугольной матрицы.  
 Например, для X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) ответ 3.

```python
# without NymPy
x = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]
n=1
asix0=len(x)
asix1=len(x[0])
diag=[x[i][i] for i in range(min(asix0,asix1)) if x[i][i]!=0]
for i in diag:
    n*=i
print(n)  

# with NymPy
x = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]])
diag=np.diagonal(x)
np.prod(diag[np.flatnonzero(diag)])
```

* __Задача 2__: Даны два вектора x и y. Проверить, задают ли они одно и то же мультимножество.  
  Например, для x = np.array([1, 2, 2, 4]), y = np.array([4, 2, 1, 2]) ответ True.

```python
# without NymPy
a = [1, 3, 2, 4]
b = [4, 1, 3, 2]
a, b = set(a), set(b)
print(a == b)

# with NymPy
x = np.array([1, 2, 2, 4])
y = np.array([4, 2, 1, 2])
np.array_equal(np.unique(x), np.unique(y))

```

* __Задача 3__: Найти максимальный элемент в векторе x среди элементов, перед которыми стоит ноль. 
 Например, для x = np.array([6, 2, 0, 3, 0, 0, 5, 7, 0]) ответ 5.

```python
# without NymPy
x = [6, 2, 0, 3, 0, 0, 5, 7, 0]
if x[-1]==0:
    x=x[:-1]
elements_befor_zero=[x[i+1] for i in range(len(x)) if x[i]==0] 
max(elements_befor_zero)

# with NymPy
x = np.array([6, 2, 0, 3, 0, 0, 5, 7,0])
if x[-1]==0:
    x=x[:-1]
max(x[np.where(x==0)[0]+1])
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
# without NymPy
x = [2, 2, 2, 3, 3, 3, 5]
count_elem=[x.count(i) for i in set(x)]
print(f'{list(set(x))}, {count_elem}')


# with NymPy
x = np.array([2, 2, 2, 3, 3, 3, 5]) 
np.unique(x, return_counts=True) 
```

* __Задача 5__: Даны две выборки объектов - X и Y. Вычислить матрицу евклидовых расстояний между объектами. Сравните с функцией scipy.spatial.distance.cdist по скорости работы.


```python
from scipy.spatial.distance import cdist
import time

a = np.array([[1, 2, 3], [4, 5, 6]])
b = np.array([[7, 8, 9], [10, 11, 12]])

# without NymPy
start_time_1 = time.time()
euclidean_distances_scipy = cdist(a, b, 'euclidean')
end_time_1 = time.time()
print(euclidean_distances_scipy)
print(f"Время работы (scipy.spatial.distance.cdist): {end_time_1 - start_time_1} сек")

# with NymPy
start_time_2 = time.time()
euclidean_distances_nympy=np.linalg.norm(a[:, np.newaxis, :] - b, axis=2)
end_time_2 = time.time()
print(euclidean_distances_nympy)
print(f"Время работы (NumPy): {end_time_1 - start_time_1} сек")

```
